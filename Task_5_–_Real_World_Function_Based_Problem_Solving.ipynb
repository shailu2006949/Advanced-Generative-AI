# **1 Social Media – Post Engagement Analyzer**


```python
def analyze_engagement(likes):
  total = 0
  for i in likes:
    total += i
    if total >= 1000:
      print("Total Likes:", total)
      print("Post Status: Viral POst")
    else:
      print("Total likes:", total)
      print("Post Status: Normal Engagement")
analyze_engagement([200, 300, 150, 400])
```

# **3. Agriculture – Rainfall Adequacy Checker**

```python
def rainfall_status(data, required):
  total = 0
  for i in data:
    total += i
  average = total // len(data)
  print("Average Rainfall:", average)
  if average >= required:
    print("Rainfall Status: Adequate Rainfall")
  else:
    print("Rainfall Status: InAdequate Rainfall")
rainfall_status([70, 75, 68, 75], 70)
```

# **4 Social Media – Duplicate Account Detection**

```python
def detect_duplicate_usernames(usernames):
    unique_usernames = set(usernames)
    if len(unique_usernames) != len(usernames):
        return "Duplicate Accounts Found: Yes"
    else:
        return "Duplicate Accounts Found: No"
user_list = ["john123", "sara456", "mike789", "john123"]
result = detect_duplicate_usernames(user_list)
print(result)
```

# **6. Agriculture – Premium Crop Price Filter**


```python
def filter_premium_crops(prices, threshold=2000):
  premium = [price for price in prices if price > threshold]
  return premium
crop_prices = [1500, 2500, 1800, 3200, 2000]
premium_crops = filter_premium_crops(crop_prices)
print("Premium Crops:", premium_crops)
```

# **7. System Monitoring - Application Health**

```python
def evaluate_systems_health(errors):
  if errors == 0:
    status = "Healthy"
  elif errors <= 5:
    status = "Minor Issues"
  else:
    status = "Critical Issues"
  return status
error_count = 7
system_status = evaluate_systems_health(error_count)
print("Error Count:", error_count)
print("System Status:", system_status)
```

# **8. Banking – Daily Transaction Limit Checker**


```python
def validate_transaction(amount, daily_limit=50000):
  if amount <= daily_limit:
    status = "Approved"
  else:
    status = "Rejected"
  return status
transaction_amount = 60000
transaction_status = validate_transaction(transaction_amount)
print("Transaction Amount:", transaction_amount)
print("Transaction_Status:", transaction_status)
```

# **9. E-Learning – Student Attendance Eligibility System**


```python
def check_attendance(attendance):
  percentage = (attendance.count("P") / len(attendance)) * 100
  eligibility = "Eligible" if percentage >= 75 else "Not Eligible"
  return percentage, eligibility
attendance = ["P","A","P","P","P","A","P","P","P","P"]
percentage, eligibility = check_attendance(attendance)
print("Attendance Percentage:", percentage)
print("Exam Eligibility:", eligibility)
```

# **10. Smart Electricity Bill Analyzer**


```python
def calculate_bill(units):
    bill = 3*min(units,100) + 5*min(max(units-100,0),100) + 7*max(units-200,0)
    usage = "Low Usage" if bill<500 else "Moderate Usage" if bill<=1500 else "High Usage"
    return bill, usage
units = 250
total, status = calculate_bill(units)
print("Units Consumed:", units)
print("Total Bill: ₹", total)
print("Usage Status:", status)
```
