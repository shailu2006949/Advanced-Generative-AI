2.Inventory Reorder Alert System

```python
def check_stock(products):
  for item, stock in products.items():
    if stock < 15:
      print(item, "-> Reorder Alert")
    else:
      print(item, "-> stock Ok")
products = {
    "Apples": 10,
    "Bananas": 25,
    "Oranges": 12,
    "Mangoes": 18
}
check_stock(products)
```

 3. Student Result Processing System


```python
def calculate_result(marks):
    total = 0
    for mark in marks:
        total += mark
    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"
student_marks = [60, 45, 70, 55]

result = calculate_result(student_marks)

print("Average Marks:", sum(student_marks)/len(student_marks))
print("Result:", result)
```

4. Cab Fare Estimator with Retry Option


```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)
    if peak.lower() == "yes":
        fare = fare + (0.25 * fare)   # 25% extra
    return fare
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")
    total_fare = calculate_fare(distance, peak)
    print("Total Fare: â‚¹", total_fare)
    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        print("Thank you for using Cab Fare Calculator ðŸš•")
        break
```

 5. Employee Attendance Eligibility

```python
def check_attendance(att):
    if len(att) == 0:
        return "No Data"

    present = 0
    for i in att:
        if i == "P":
            present += 1
    percent = (present / len(att)) * 100
    return "Eligible" if percent >= 75 else "Not Eligible"
attendance = ["P","A","P","P","P"]
print(check_attendance(attendance))
```

6. Password Strength Checker


```python
def check_password(pwd):
  if len(pwd) < 8:
    return "Weak Password"
  has_digit = False
  has_special = False
  for ch in pwd:
    if ch.isdigit():
      has_digit = True
    if ch in "@#$":
      has_special = True
  if has_digit and has_special:
    return "Strong Password"
  else:
    return "Weak Password"
print(check_password("Pass@123"))
```
