1 .Employee Performance Bonus Eligibility


```python

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest_score = max(employees.values())
top_performers = [name for name, score in employees.items() if score == highest_score]
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", highest_score, ")")
```

 2 .Search Query Keyword Analysis


```python
import string
sentence = "Buy mobile phone buy phone online"
sentence = sentence.lower()
sentence = sentence.translate(str.maketrans('', '', string.punctuation))
words = sentence.split()
frequency = {}
for word in words:
  frequency[word] = frequency.get(word, 0) + 1
result = {word: count for word, count in frequency.items() if count > 1}
print(result)
```

3.Sensor Data Validation


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = [(i, v) for i, v in enumerate(sensor_readings) if v % 2 == 0]
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

4 .Email Domain Usage Analysis


```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
total = len(emails)
count = {}

for e in emails:
    d = e.split("@")[1]
    count[d] = count.get(d, 0) + 1

for d in count:
    print(d + ":", str(int(count[d] / total * 100)) + "%")
```

5.Sales Spike Detection


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
avg = sum(sales) / len(sales)
for i, s in enumerate(sales):
  if s > avg * 1.3:
    print("Day", i+1, ":", s)
```

6.Duplicate User ID Detection


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
count = {}
for u in user_ids:
  count[u] = count.get(u, 0) + 1
for u in count:
  if count[u] > 1:
    print(u, "->", count[u], "times")
```
